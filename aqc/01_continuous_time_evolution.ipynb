## Continuous-Time Evolution in Quantum Systems

We have seen that in quantum computing, the evolution of a closed system is described by a unitary transformation.

That is, the state of the system $\ket{\psi(t)}$ at time $t$ can be expressed as

\begin{equation}
\ket{\psi(t)} = U(t,t_0)\ket{\psi(t_0)},
\end{equation}

where $U(t,t_0)$ is a unitary operator that depends on times $t$ and $t_0$.

An equivalent way of stating this fact is the Schrödinger equation. The Schrödinger equation provides a continuous-time description of the system while through Unitary operators, we have a discrete-time description:

$$
H\ket{\psi(t)} = i\hbar \frac{d}{d t} \ket{\psi(t)},
$$

where $\hbar$ is a physical constant known as Planck’s constant and $\frac{d}{d t}$ is the derivative. In
practice, it is common to absorb the factor $\hbar$ into $H$, the Hamiltonian describing the system.


The Schrödinger equation is a differential equation, whose solution is given by

$$
\ket{\psi(t)} = \exp \left(-iH\cdot(t-t_0)\right) \ket{\psi(t_0)}.
$$

One can prove that $e^{-iH(t-t_0)}$ is always unitary and any unitary can always be expressed as $e^{-iK}$ for some Hermitian operator $K$. This proves the equivalence between the two alternative ways of viewing time evolution in quantum systems.

In other words, if the state of a system at time $t_0$ is $\ket{\psi(t_0)}$, its state at time $t$ is given by $e^{-iH(t-t_0)}\ket{\psi(t_0)}$ and the corresponding unitary transformation is given by

$$
U(t,t_0) = \exp \left(-iH\cdot(t-t_0)\right).
$$


Note that if $\ket{\psi}$ is an eigenstate of $H$, it only acquires a phase when it is evolved by Hamiltonian $H$, as it is not possible to transition between eigenstates.

### Matrix exponentials

For a given matrix $A$, the matrix exponential $\exp(A)$ is defined as $\exp(A) = \sum_{k=0}^\infty \frac{A^k}{k!}$.

For instance, for $A = \begin{pmatrix} 1 & 0 \\ 0 & 0\end{pmatrix}$, let us compute it step by step:

\begin{align}
\exp(A) &= I + A + \frac{A^2}{2!} + \frac{A^3}{3!} + \dots \\
&= \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix} + \begin{pmatrix} 1 & 0 \\ 0 & 0 \end{pmatrix} + \frac{1}{2!}\begin{pmatrix} 1 & 0 \\ 0 & 0 \end{pmatrix} + \dots \\
&= \begin{pmatrix} 1 + 1 + \frac{1}{2!} + \frac{1}{3!} + \dots & 0 \\ 0 & 1 + 0 + 0 + \dots \end{pmatrix} \\
&= \begin{pmatrix} \sum_{k=0}^\infty \frac{1^k}{k!} & 0 \\ 0 & 1 \end{pmatrix} \\
&= \begin{pmatrix} e & 0 \\ 0 & 1 \end{pmatrix}.
\end{align}

Hermitian matrices are diagonalizable, i.e., if $A$ is Hermitian, $A$ can be written as $A= PDP^{-1}$ where $P$ is a unitary matrix. Using this fact, we can show that for a Hermitian matrix $A$:

$$
\exp(A) = \sum_{k=0}^\infty \frac{(PDP^{-1})^k}{k!} = \sum_{k=0}^\infty\frac{PD^kP^{-1}}{k!} = P\exp(D)P^{-1}.
$$

For a diagonal matrix
$$D = \begin{pmatrix}
d_1 & 0 & \cdots & 0 \\
0 & d_2  & \cdots & 0 \\
\vdots & \vdots & \ddots & \vdots \\
0 & 0  & \cdots & d_n
\end{pmatrix},$$

one can show that
$$ \exp(D) = \begin{pmatrix}
\exp(d_1) & 0  & \cdots & 0 \\
0 & \exp(d_2)  & \cdots & 0 \\
\vdots & \vdots & \ddots & \vdots \\
0 & 0  & \cdots & \exp(d_n)
\end{pmatrix}.
$$

Note that $\exp(-iDt)$ is unitary,  since its diagonal entries are all with modulus 1.


Let us go through an example and compute $\exp(-iHt)$ for 
$$
H = \begin{pmatrix} 2 & 1 \\ 1 & 2 \end{pmatrix}.
$$

Step 1: Eigenvalues. 

The characteristic equation is
$$
\det(H - \lambda I) = \det \begin{pmatrix} 2-\lambda & 1 \\ 1 & 2-\lambda \end{pmatrix} = (2-\lambda)^2 - 1 = 0,
$$
which gives
$$
\lambda_1 = 1, \quad \lambda_2 = 3.
$$

Step 2: Eigenvectors.

For $\lambda_1 = 1$:
$$
(H - I)v = \begin{pmatrix} 1 & 1 \\ 1 & 1 \end{pmatrix} \begin{pmatrix} x \\ y \end{pmatrix} = 0 \implies x + y = 0 \implies v_1 = \begin{pmatrix} 1 \\ -1 \end{pmatrix}.
$$

For $\lambda_2 = 3$:
$$
(H - 3I)v = \begin{pmatrix} -1 & 1 \\ 1 & -1 \end{pmatrix} \begin{pmatrix} x \\ y \end{pmatrix} = 0 \implies -x + y = 0 \implies v_2 = \begin{pmatrix} 1 \\ 1 \end{pmatrix}.
$$

Step 3: Diagonalization.

We have
$$
P = \begin{pmatrix} 1 & 1 \\ -1 & 1 \end{pmatrix}, \quad
D = \begin{pmatrix} 1 & 0 \\ 0 & 3 \end{pmatrix}, \quad
P^{-1} = \frac{1}{2} \begin{pmatrix} 1 & -1 \\ 1 & 1 \end{pmatrix}.
$$

Step 4: Matrix Exponential. 

Then
\begin{align*}
\exp(-iHt) &= P \, \exp(-iDt) \, P^{-1} \\
&= \begin{pmatrix} 1 & 1 \\ -1 & 1 \end{pmatrix}
\begin{pmatrix} e^{-i t} & 0 \\ 0 & e^{-i 3 t} \end{pmatrix}
\frac{1}{2} \begin{pmatrix} 1 & -1 \\ 1 & 1 \end{pmatrix} \\
&= \frac{1}{2} 
\begin{pmatrix} e^{-i t} + e^{-i 3 t} & -e^{-i t} + e^{-i 3 t} \\ -e^{-i t} + e^{-i 3 t} & e^{-i t} + e^{-i 3 t} \end{pmatrix}.
\end{align*}

### Matrix Exponentials for Pauli operators $P$

Let us check $\exp(-iPt)$ for Pauli operators $\sigma_x, \sigma_y$, and $\sigma_z$. Note that $P^2 = I$ for each Pauli operator $P$. Therefore, we can make use of the following fact:

$$
\exp(iAx) =\cos(x)I + i\sin(x)A
$$
for real $x$ and matrix $A$ such that $A^2 = I$.

Using this identity,

\begin{align}
\exp({-i P t}) &= \cos(t) I - i \sin(t) P.
\end{align}

Now, let us check it for each Pauli matrix.

1) $\sigma_x$

\begin{align}
\exp({-i \sigma_x t}) &= \cos(t) I - i \sin(t) \sigma_x \\
                  &= \cos(t) \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix} - i \sin(t) \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix} \\
                  &= \begin{pmatrix}
                  \cos(t) & -i \sin(t) \\
                  -i \sin(t) & \cos(t)
                  \end{pmatrix} \\
                  &= R_x(2t).
\end{align}

This represents a rotation around the $ x $-axis by angle $2t$.

2) $ \sigma_y $

\begin{align}
\exp({-i \sigma_y t}) &= \cos(t) I - i \sin(t) \sigma_y \\
                  &= \cos(t) \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix} - i \sin(t) \begin{pmatrix} 0 & -i \\ i & 0 \end{pmatrix} \\
                  &= \begin{pmatrix}
                  \cos(t) & \sin(t) \\
                  -\sin(t) & \cos(t)
                  \end{pmatrix}\\
                  &= R_y(2t).
\end{align}

This represents a rotation around the $ y $-axis by angle $2t$.

3) $ \sigma_z $

\begin{align}
\exp({-i \sigma_z t}) &= \cos(t) I - i \sin(t) \sigma_z \\
                  &= \cos(t) \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix} - i \sin(t) \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix} \\
                  &= \begin{pmatrix}
                  e^{-it} & 0 \\
                  0 & e^{it}
                  \end{pmatrix}\\
                  &= R_z(2t).
\end{align}

This represents a rotation around the $ z $-axis by angle $2t$.


Thus, the matrix exponentials of the Pauli matrices give the rotation matrices around the respective axes $ x, y$, and $ z$.

### Qiskit tools for time evolution

Qiskit provides built-in tools for simulating the unitary 
time evolution $\exp(-iHt)$ of a quantum system. `HamiltonianGate` takes a Hamiltonian matrix $H$ and a time $t$ as inputs and directly computes the unitary $\exp(-iHt)$. 


For example, to create the evolution under $H = X$ for time $t=2$:


```python
import numpy as np
from qiskit.circuit.library import HamiltonianGate
from qiskit.quantum_info import SparsePauliOp

t = 2
H = SparsePauliOp(['X'], coeffs=[1])
hgate = HamiltonianGate(H, time=t)
```

`HamiltonianGate` accepts any Hermitian matrix as input, not necessarily expressed through Pauli terms.

`PauliEvolutionGate` uses the **Lie-Trotter product formula** to approximate the time evolution. For a Hamiltonian that is a sum of Pauli terms, $H = \sum_k h_k$, the exact evolution $\exp(-iHt)$ is approximated by implementing each Pauli term separately:

$$\exp\left(-i\sum_k h_k t\right) =\lim_{n \rightarrow \infty} \left(\prod_k \exp\left(\frac{-ih_k t}{n}\right)\right)^n.$$

By default, `PauliEvolutionGate` uses a single repetition $n=1$, so each Pauli term is implemented as a separate gate applied consecutively. This is an approximation and is not exact in general.

`HamiltonianGate`, on the other hand, computes the matrix exponential directly and gives the exact unitary. Note that `PauliEvolutionGate` is also not exact for Hamiltonians acting on more than 2 qubits.

## Time-Dependent Hamiltonian


Hamiltonian may change in time and vary according to some parameters. In this case, the system evolves according to Schrödinger’s
equation with a time-dependent Hamiltonian:





$$
H(t)\ket{\psi(t)} = i\hbar \frac{d}{d t} \ket{\psi(t)}.
$$



The evolution of the system is much more complicated with time-dependent Hamiltonian. In this case, the corresponding unitary transformation is given by

$$
U(t,t_0) = \exp \left({-\frac{i}{\hbar} \int_{t_0}^t dt' H(t')}\right ) .
$$

The quantum adiabatic theorem explains the behaviour of the system when the Hamiltonian changes sufficiently slowly.

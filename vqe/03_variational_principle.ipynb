# Variational Principle



For a Hamiltonian $H$ with eigenstates $|\psi_\lambda\rangle$ and corresponding eigenvalues $\lambda$, the variational principle states that for any normalized trial state $|\psi\rangle$, the following is true:

$$
E_0 \leq \langle \psi | H | \psi \rangle,
$$

where $E_0$ is the ground-state energy (the lowest eigenvalue of $H$).

Now let us understand why this is the case. 

Since any state $\ket{\psi}$ can be expressed as a linear combination of the eigenstates  $\ket{\psi}\sum_{\lambda} c_{\lambda} \ket{\lambda}$ such that $\sum_{\lambda} |c_\lambda|^2 = 1$,
\begin{align*}
\langle \psi | H | \psi \rangle  &=   \bra{\sum_\lambda c_\lambda^* \psi_\lambda}H \ket{ \sum_\mu c_\mu \psi_\mu} \\
& = \sum_{\lambda, \mu} c_\lambda^* c_\mu \langle \psi_\lambda | H | \psi_\mu \rangle. \\
\end{align*}

Now since $H |\psi_\mu\rangle = \mu |\psi_\mu\rangle$, we get:
$$
\langle \psi | H | \psi \rangle  = \sum_{\lambda, \mu} c_\lambda^* c_\mu  \mu \langle \psi_\lambda | \psi_\mu \rangle.
$$

Since the eigenstates are orthonormal, all terms vanish except for the case $\mu = \lambda$.

$$
\langle \psi | H | \psi \rangle= \sum_{\lambda} \lambda |c_\lambda|^2 
$$


Since $E_0$ is the lowest eigenvalue of $H$, it follows that:

$$
E_0 \sum_{\lambda} |c_\lambda|^2 \leq \sum_{\lambda} \lambda|c_\lambda|^2  = \langle \psi | H | \psi \rangle.
$$

Since $\sum_{\lambda} |c_\lambda|^2 = 1$, this simplifies to:

$$
E_0 \leq \langle \psi | H | \psi \rangle.
$$

### What does it signify?

This proves that $\langle \psi | H | \psi \rangle$ for any state $|\psi\rangle$ is always greater than or equal to the ground-state energy $E_0$. In other words, any random guess of a quantum state will provide an upper bound on the ground state energy. This principle forms the basis for the Variational Quantum Eigensolver, which aims to approximate $E_0$ by optimizing over different trial states.





The quantity $\langle \psi | H | \psi \rangle$ is called the expectation value of $H$ with respect to $\psi$. We will explain why it is the expectation value of $H$ in the following notebook.

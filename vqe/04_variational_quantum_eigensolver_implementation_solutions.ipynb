### Task 1

For $H=Z$, compute $P_\lambda$.

### Solution

Eigenvalues of $Z $are $-1$ and $1$, with the corresponding eigenvectors $\ket {\psi_{-1}} = \myvector{0 \\ 1}$ and $\ket {\psi_{1}} = \myvector{1 \\ 0}$, respectively. In this case,

$$
P_{-1} = \myvector{ 0&0\\ 0 & 1} \text{ and } P_{1} = \myvector{ 1&0\\ 0 & 0}.
$$

### Task 2

Compute the expectation value of $Z$ operator with respect to the state $\ket{\psi} = \frac{1}{\sqrt 3} \ket 0 + \frac{\sqrt 2}{\sqrt 3} \ket 1$. Use the formula $
\bra{\psi}H\ket{\psi} =  \sum_\lambda  \lambda \bra{\psi} P_\lambda\ket{\psi}  = \sum_\lambda  \lambda p(\lambda) = E(H)$.

Also compute $\bra{\psi}H\ket{\psi}$ directly and compare.

### Solution


Note that $P_{-1}\ket \psi = \myvector{0 \\ \frac{\sqrt 2}{\sqrt 3} }$ and $P_{1}\ket \psi = \myvector{ \frac{ 1}{\sqrt 3} \\ 0}$. Therefore,

$$
p(-1) = \bra{\psi}P_{-1}\ \ket{\psi}=  \frac{2}{3} \text{ and } p(1) =  \bra{\psi}P_{1}\ \ket{\psi} = \frac{1}{3}.
$$


Hence we get,

$$
E(H) = p(-1)\cdot -1 + p(1) \cdot 1 = -\frac{2}{3}+\frac{1}{3} = -\frac{1}{3}
$$


Now let's compute directly.

\begin{align}
 \bra{\psi} Z\ket{\psi} &=
 \left( \frac{1}{\sqrt 3}  \langle 0| +\frac{\sqrt 2}{\sqrt 3}  \langle 1| \right)  \left(  \frac{1}{\sqrt 3} |0\rangle - \frac{\sqrt 2}{\sqrt 3}  |1\rangle  \right) \\
 &= \frac{1}{3}  \langle 0|0\rangle -   \frac{\sqrt 2}{ 3} \langle 0|1\rangle +  \frac{\sqrt 2}{ 3} \langle 1|0\rangle -  \frac{2}{ 3}  \langle  1|1\rangle  \\
 &= \frac{1}{3} -  \frac{2}{3}= -\frac{1}{3}
\end{align}

### Task 9

Using the ansatz given below, run VQE to estimate the ground state energy of the Hamiltonian $H = 2Y_0Y_1 + Z_0X_1 - X_0Z_1$.  Compute the ground state energy exactly to verfiy your findings.




```python
from qiskit import  QuantumCircuit
from qiskit.circuit import Parameter

def ansatz():
    qc = QuantumCircuit(2,1)
    theta_0 = Parameter('theta_0')
    theta_1 = Parameter('theta_1')
    qc.rx(theta_0,0)
    qc.cx(0,1)
    qc.ry(theta_1,1)
    return qc
```

### Solution

To start with, find the lowest eigenvalue of our operator using `scipy` to verify your findings later on.


```python
import numpy as np
from qiskit.quantum_info import SparsePauliOp

H = SparsePauliOp(["YY", "ZX", "XZ"], [2.0, 1.0, -1.0])
eigenvalues = np.linalg.eigvals(H)

print("Eigenvalues:", eigenvalues)
print("Ground state energy (smallest eigenvalue):", np.min(eigenvalues).real)
```

Complete the following function, that returns the necessary quantum circuit for measurement with respect to the given Hamiltonian `op`.

```python
# op will be one of Y0Y1, Z0X1, X0Z1
def measurement_circuit(op):
    qc = QuantumCircuit(2)
    #p represents the Pauli string in the circuit and i is its index
    for i, p in enumerate(op):
        if str(p) == 'X':
            qc.h(i)
        elif str(p) == 'Y':
            qc.sdg(i)
            qc.h(i)
    #We can have this since all observables act on both qubits non-trivially
    qc.cx(0,1)
    return qc
```

Complete the function below, that computes the expectation value of the ansatz with respect to $H$, based on the given parameters.

```python
from qiskit.circuit import Parameter
from qiskit import  QuantumCircuit, transpile
from qiskit.circuit.library import StatePreparation
from qiskit_aer import AerSimulator

#The first parameters should be always the parameters
def expectation_value(params, H):
    e = 0
    #For each term in the Hamiltonian
    for op in H:
        #Construct the ansatz
        qc = ansatz()
        #Asign the parameters
        bc = qc.assign_parameters(params)
        #Add the measurement circuit and measure
        bc = bc.compose(measurement_circuit(op.paulis[0]))
        bc.measure(1,0)
        #Transpile the circuit for convenience
        bc = transpile(bc, basis_gates=['u', 'cx'])
        #Run the circuit and get the counts
        shots = 10000
        job = AerSimulator().run(bc,shots=shots)
        counts = job.result().get_counts(bc)
        #Compute expectation for each term in the Hamiltonian and sum them up
        e += op.coeffs[0]*(-1*counts.get("1",0) / shots + 1*counts.get("0",0) / shots)
    return e.real
```

We set some initial values and check the function's value at the initial parameters.

```python
# Set initial parameters
params = [0.5, 0.5]
H = SparsePauliOp(["YY", "ZX", "XZ"], [2.0, 1.0, -1.0])
e = expectation_value(params, H)
print("Initial expectation value estimation:", e)
```

Use a classical optimizer to find the parameters, which minimizes the function.


```python
from scipy.optimize import minimize

#Note that for x0 we pass some initial parameters 
#and for args we pass the second argument of the expectation value function which is the Hamiltonian H
result = minimize(expectation_value, x0 = params, args=(H,), method='COBYLA', options = {"maxiter" : 10000})
print("Optimized parameters:", result.x)
print("Expectation value at optimized parameters:", expectation_value(result.x, H))
```

VQE gives us an upper bound on the ground state energy $\approx-1$. This is quite far from the actual ground state energy, which is $-4$.

One can try different classical optimization algorithms to improve it or to increase the number of iterations of the classical optimization algorithm. But in the example above, our ansatz is not expressive enough.

### Solution



```python
from qiskit.quantum_info import Operator
import numpy as np

XYY = Operator.from_label("XYY")
ZZI = Operator.from_label("ZZI")
YXY = Operator.from_label("YXY")
H = 3*XYY + 2*ZZI - -4*YXY

assert(np.allclose(H.data, H.data.conj().T))

print(H)

```

### Solution

```python
from qiskit.quantum_info import SparsePauliOp

def random_hamiltonian(n_qubits, n_terms):
    
    paulis = ["I", "X", "Y", "Z"]
    labels = []
    coeffs = []
    for _ in range(n_terms):
        label = "".join(np.random.choice(paulis, size=n_qubits))
        coeff = np.random.uniform(-1, 1)
        labels.append(label)
        coeffs.append(coeff)
    H = SparsePauliOp(labels, coeffs)

    assert np.allclose(H.to_matrix(), H.to_matrix().conj().T)
    return H
    
```

```python
random_hamiltonian(4,3)
```

### Solution

```python
import numpy as np

H  = random_hamiltonian(4,5)
eigenvalues, eigenvectors = np.linalg.eigh(H.to_matrix())
ground_state_energy = np.min(eigenvalues)
ground_state = eigenvectors[np.argmin(eigenvalues)]

for e in eigenvalues:
    assert np.isclose(e.imag, 0)

print("Ground state energy:", ground_state_energy)
print("Ground state:", ground_state)
```

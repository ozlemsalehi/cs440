```python
import numpy as np  

def generate_random_state(num_qubits):
    random_state =np.random.rand(2**num_qubits) + 1j * np.random.rand(2**num_qubits)  # Random complex numbers
    random_state /= np.linalg.norm(random_state)  # Normalize the state to make it a unit vector
    return random_state

def compute_expectation_value(state, operator):
    energy = state.conj().T @ operator @ state
    #Alternatively energy = np.vdot(state, operator @ state)
    return energy

H = Operator.from_label("Z")
energies = []
psis = []
for i in range(1000):
    psi = generate_random_state(1)
    energy = compute_expectation_value(psi, H.data)
    energies.append(energy.real) #To get rid of small imaginary parts due to numerical errors
    psis.append(psi)

min_ind = np.argmin(energies)
print("Minimum energy:", energies[min_ind])
print("Corresponding state:", psis[min_ind])
```

```python
from qiskit.quantum_info import SparsePauliOp
import numpy as np

H = SparsePauliOp(["XY", "YZ"], [4, -1])
eigenvalues, eigenvectors = np.linalg.eig(H.to_matrix())
ground_state_energy = np.min(eigenvalues).real #To get rid of small imaginary parts due to numerical errors
ground_state = np.round(eigenvectors[np.argmin(eigenvalues)], decimals=2)


print("Ground state energy:", ground_state_energy)
print("Ground state:", ground_state)
```

```python
import numpy as np  
from qiskit.quantum_info import Statevector

def generate_random_state(num_qubits):
    random_state =np.random.rand(2**num_qubits) + 1j * np.random.rand(2**num_qubits)  # Random complex numbers
    random_state /= np.linalg.norm(random_state)  # Normalize the state to make it a unit vector
    return random_state

energies = []
psis = []
for i in range(1000):
    psi = generate_random_state(2)
    energy = Statevector(psi).expectation_value(H)
    energies.append(energy.real)
    psis.append(psi)

min_ind = np.argmin(energies)
min_energy = energies[min_ind]
corresponding_state = psis[min_ind]
print("Minimum energy:", min_energy)
print("Corresponding state:", corresponding_state)
```
